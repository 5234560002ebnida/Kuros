```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

import pandas as pd

Url_B_reservation="https://raw.githubusercontent.com/5234560002ebnida/nonfictioner/main/Hotel%20Reservations.csv"

### 자료 불러서 df라고 하자.

df = pd.read_csv(Url_B_reservation)

### 자료가 어떻게 생겼나?
df.head()
```

```python
### 뒤에서 10개를 볼까요?

df.tail(n=10)
```

```python
### 대충 정보를 확인하자.

df.info()
```

```python
X= df.iloc[:,0:17]
y= df.iloc[:,17]
```

```python
k=5
```

```python
kfold = KFold(n_splits=k, shuffle=True)
```

```python
acc_score = []
```

```python
df.drop('Booking_ID', axis=1, inplace=True)
df.head()
```

```python
df['room_type_reserved'].value_counts()
```

```python
encoder = LabelEncoder()
```

```python
# Turning the object values to meaningful numbers using LabelEncoder
df[['type_of_meal_plan','room_type_reserved',
      'market_segment_type']] = df[[
          'type_of_meal_plan','room_type_reserved','market_segment_type',]].apply(encoder.fit_transform)

mapping = {'Not_Canceled':0,'Canceled':1}
df['booking_status'] = df['booking_status'].map(mapping)

df.head()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def model_fn():
  model = Sequential()
  model.add(Dense(24, input_dim=18, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  return model
```

```python
for train_index, test_index in kfold.split(X):
  X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model = model_fn()
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)

accuracy = model.evaluate(X_test, y_test)[1]
acc_score.append(accuracy)
```

```python
avg_acc_score= sum(acc_score) / k
```

```python
print('정확도',acc_score)
print('정확도 평균: ', avg_acc_score)
```
