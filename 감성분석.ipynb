```python
from transformers import pipeline

senti = pipeline('sentiment-analysis')
```

```python
senti.model
```

```python
##로빈 윌리엄스, 죽은 시인의 사회

print(senti(["Carpe Diem, Seize the day. Make your lives extraordinary."]))
```
